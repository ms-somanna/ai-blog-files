```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.graph_objects as go
from matplotlib.image import imread
```

```python
# Data
x = np.arange(0,360,1)
y = np.sin(np.deg2rad(x))

# Noise
noise = np.random.normal(0,0.1, 360)
#noise = np.concatenate([noise, np.zeros(300)])
#np.random.shuffle(noise)

# Data with noise
y_noised = y + noise
```

```python
# Plot each array
fig = make_subplots(rows=1, cols=5,
                    print_grid=False,
                    column_widths=[0.25, 0.125, 0.25, 0.125, 0.25])

# Plot data D
fig.add_trace(
    go.Scatter(x=x,y=y, marker_color='red'),
    row=1, col=1,
)

# Plot symbol: "+"
fig.add_trace(
    go.Scatter(x=[2], y=[0], mode='markers', marker_symbol='cross', marker_size=30, marker_color='black'),
    row=1, col=2,
)

# Plot noise N
fig.add_trace(
    go.Histogram(x=noise, marker_color='limegreen'),
    row=1, col=3
)

# Plot symbol: "-->"
fig.add_trace(
    go.Scatter(
        x=[0],  # X coordinate for the text
        y=[0],  # Y coordinate for the text
        mode='text',
        text=['➔'],
        textfont=dict(size=60)
    ),
    row=1, col=4,
)

# Plot data with noise added
fig.add_trace(
    go.Scatter(x=x,y=y_noised, marker_color='gold'),
    row=1, col=5
)
fig.update_layout(height=300, width=1000,
                  showlegend=False,
                  paper_bgcolor = 'rgba(0, 0, 0, 0)',
                  plot_bgcolor = 'rgba(0, 0, 0, 0)')
fig.update_yaxes(visible=False, showticklabels=False)
fig.update_xaxes(visible=False, showticklabels=False)
fig.show()
```

```python
# Plot each array
fig = make_subplots(rows=1, cols=3,
                    subplot_titles=['Original Data','', 'Gaussian Noise', 'Data with Gaussian Noise Added'])

# Plot data D
fig.add_trace(
    go.Scatter(x=x,y=y, marker_color='blue'),
    row=1, col=1,
)

# Plot noise N
fig.add_trace(
    go.Histogram(x=noise),
    row=1, col=2
)

# Plot data with noise added
fig.add_trace(
    go.Scatter(x=x,y=y_noised, marker_color='blue'),
    row=1, col=3
)
fig.update_layout(height=400, width=1000,
                  title_text="Fig-1: Effect of Gaussian Noise on Sinusoidal Data",
                  showlegend=False)
fig.show()
```
